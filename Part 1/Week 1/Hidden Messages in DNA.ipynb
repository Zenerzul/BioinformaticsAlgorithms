A Faster Frequent Words Approach

```python
def FrequencyTable(Text, k):
    '''
    Function to create dictionary of all k-mers and their counts in chosen Text
    '''
    freqMap = dict()

    for i in range(len(Text) - k + 1):
        Pattern = Text[i:i+k]
        if Pattern not in freqMap:
            freqMap[Pattern] = 1
        else:
            freqMap[Pattern] += 1
    return freqMap
```

```python
def SortPatternsDict(Dictionary):
    '''
    Function return dictionary sorted by value.
    '''

    sortedDict = {
        key: value for key, value in sorted(Dictionary.items(), key=lambda item: item[1], reverse=True)
    }
    return sortedDict
    
    
def MaxFreq(Dictionary: dict):
    '''
    Function to return max value from dictionary
    '''

    maxValue = max(
        Dictionary, key=Dictionary.get
    )
    return Dictionary[maxValue]
```

```python
def BetterFrequentWords(Text, k):
    '''
    Function to find the most common k-mers of certain length in text
    '''
    
    frequentPatterns = list()
    freqMap = FrequencyTable(Text, k)

    max_freq = MaxFreq(freqMap)

    for pattern in freqMap:
        if freqMap[pattern] == max_freq:
            frequentPatterns.append(pattern)

    return frequentPatterns
```

```python
def ReverseComplement(String):
    '''
    Function to return reverse complement of given DNA strand at 5->3 direction
    '''

    nucleotidesDict = {
        "A": "T", "T": "A", "G": "C", "C": "G"
    }

    reverseComplement = str()
    for i in String:
        reverseComplement += nucleotidesDict[i]
    
    return reverseComplement[::-1]
```

```python
def PatternMatching(Pattern, String):
    '''
    Function to output pattern positions in string
    '''

    totalCoordinate = 0
    positions = list()

    for i in range(PatternCount(String, Pattern)):
        '''
        We need to find match, then move to the matching index plus one to exclude repeating match, but to allow overlapping match
        '''
        
        startCoordinate = String.index(Pattern)
        totalCoordinate += startCoordinate
        positions.append(totalCoordinate)

        String = String[startCoordinate+1:]
        totalCoordinate += 1

    print(*positions)

```

Для паттерна и его комплементарной последовательности находим участок, где оба варианта встречаются с повышенной плотностью, нежели чем в остальном геноме. Значит мы хотим искать участки генома, где эти паттерны часто встречаются (clumps) 

```python
def FindClumps(Text, k, L, t):
    '''
    Function to find clumps of k-mers in window of size L more than t times
    '''

    Patterns = set()
    n = len(Text)

    for i in range(n - L):
        Window = Text[i:L+i]
        # print(Window)
        freqMap = FrequencyTable(Text=Window, k=k)
        # print(SortPatternsDict(freqMap))
        
        for key in freqMap:
            if freqMap[key] >= t:
                Patterns.add(key)
    
    return Patterns
```
